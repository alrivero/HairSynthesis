```python
import os
import json
import pandas as pd
```

```python
def read_json(json_path) -> dict:
    with open(json_path, 'r') as f:
        data = json.load(f)
    
    return data
```

```python
ROOT_DIR = "/gpfs/projects/CascanteBonillaGroup/thinguyen/storage/smirk/logs/smirkhair"

result_dir_map = {
    'normmag_depthfrozen': "20251203_011904_normmag_depthfrozen",
    'normmag_nodepth': "20251203_012259_normmag_nodepth",
    'normmag_local_depthfrozen': "20251203_113333_normmag_local_depthfrozen",
    'normmag_local_nodepth': "20251203_124425_normmag_local_nodepth"
}

out_results = []
for k, v in result_dir_map.items():
    _data_blur = read_json(os.path.join(ROOT_DIR, v, 'stress_test_blur', 'cosine_stats.json'))
    # _data_noise_01 = read_json(os.path.join(ROOT_DIR, v, 'stress_test_noise_0.1', 'cosine_stats.json'))
    _data_noise_05 = read_json(os.path.join(ROOT_DIR, v, 'stress_test_noise_0.5', 'cosine_stats.json'))

    out_results.append({
        'model': k,
        'sim_blur': _data_blur['sim_blur_mean'],
        # 'sim_noise_0.1': _data_noise_01['sim_blur_mean'],
        'sim_noise_0.5': _data_noise_05['sim_blur_mean'],
        'random': _data_blur['sim_random_mean']
    })

df = pd.DataFrame(out_results).set_index('model')
```

```python
df.to_csv('stress_test.csv')
```
